```python
import yfinance as yf
import pandas as pd 
```

```python
#define Ticker

tesla=yf.Ticker("TSLA")

```

```python
data= tesla.history(period="max")

# reset index , and display the first five rows

data.reset_index(inplace=True)

data.head()
```

```python
import matplotlib.pyplot as plt

import requests
from bs4 import BeautifulSoup
import plotly.graph_objects as go
from plotly.subplots import make_subplots
```

```python
#define Ticker

gme=yf.Ticker("GME")

```

```python
game=gme.history(period="max")

game.reset_index(inplace=True)

game.head()
```

```python
def make_graph(df1, df2):

    fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(24,8))

    axes[0].plot(df1['Date'], df1['Close'])
    axes[0].set_xlabel('Date')
    axes[0].set_ylabel('Closing price')
    axes[0].set_title('Tesla stock data')
    axes[0].set_xlim(pd.to_datetime('2003'), pd.to_datetime('2022'))


    axes[1].plot(df2['Date'], df2['Open'])
    axes[1].set_xlabel('Date')
    axes[1].set_ylabel('Closing price')
    axes[1].set_title(' ')
    axes[1].set_xlim(pd.to_datetime('2003'), pd.to_datetime('2022'))
    
    
    
    plt.tight_layout()

    plt.show()
```

```python
make_graph(data,data)
```
